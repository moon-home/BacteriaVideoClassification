```python
import os
from os import listdir
from os.path import isfile, join
```

```python
source_path = './Y1'
out_path = './VidAVI'

onlyfiles = [f for f in listdir(source_path) if isfile(join(source_path, f))]

for InputVidName in onlyfiles:
    if InputVidName[-3:] == 'MP4':
        infile = InputVidName
        outfile = InputVidName[:-4]+'.avi'
        infile_path = join(source_path, infile)
        outfile_path = join(out_path, outfile)
        cmdline = 'ffmpeg -i ' + infile_path + ' -b 5000k ' + outfile_path
        print(cmdline)
        os.system(cmdline)
```

```python
import cv2
vidcap = cv2.VideoCapture('tanmoy.avi')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
