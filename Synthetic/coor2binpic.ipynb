```python
def txt2list(txt_name):
    f = open(txt_name,"r")
    lines = [line for line in f.readlines()]
    f.close()
    xyt_list = []
    for line in lines:
        line_str_list = re.split('\t|\n',line)
        x = float(line_str_list[0])
        y = float(line_str_list[1])
        t = float(line_str_list[2])
        xyt_list.append([x, y, t])
    return xyt_list
    
def list2dic(xyt_list):
    t_dict = {}
    for x, y, t in xyt_list:
        if t in t_dict.keys(): 
            t_dict[t].append([x,y])
        else:
            t_dict[t] = [[x, y]]
    return t_dict

def dic2pic(t_dict, num_pic = 10, sparse = 20, folder_name = "synthetic"):
    n = 0
    for t in t_dict.keys():
        if n> (num_pic-1)*sparse: break
        if n%sparse != 0: 
            #print("n%sparse != 0, n:", n)
            n+=1
            continue
        #print("n%sparse == 0, n:", n)  
        n+=1
        x=[];y=[]
        for xx, yy in t_dict[t]:
            x.append(xx)
            y.append(yy)
            
        figure(figsize=(25, 15), dpi=80, edgecolor='k')
        ax = plt.gca()
        ax.axes.get_xaxis().set_visible(False)
        ax.axes.get_yaxis().set_visible(False)
        ax.set_facecolor('gray')
        plt.plot(x,y, 'wo', markersize=5)
        
        file_name = folder_name + "_" + str(t) + ".png"
        print(file_name)
        plt.savefig(file_name, bbox_inches="tight")
        plt.show()
        plt.close()
   
def txt2pic(txt_name):
    xyt_list = txt2list(txt_name)
    t_dict = list2dic(xyt_list)
    dic2pic(t_dict)
```
